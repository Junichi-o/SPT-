```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint, quad
from scipy.interpolate import interp1d

# ---  ---
Omega_m = 0.3
sigma8_0 = 0.815
rho_crit0 = 3 * (70 * 1e3 / 3.086e22)**2 / (8 * np.pi * 6.6743e-11)
rho_m = Omega_m * rho_crit0

# ---  ---
obs_Nz = [
    {"z": 0.1, "N": 90, "err": 15},
    {"z": 0.2, "N": 150, "err": 20},
    {"z": 0.3, "N": 180, "err": 20},
    {"z": 0.4, "N": 160, "err": 18},
    {"z": 0.5, "N": 110, "err": 15},
    {"z": 0.6, "N": 70, "err": 12},
    {"z": 0.7, "N": 40, "err": 10}
]

obs_fs8 = [
    {"z": 0.02, "val": 0.428, "err": 0.046},
    {"z": 0.15, "val": 0.49,  "err": 0.145},
    {"z": 0.32, "val": 0.384, "err": 0.095},
    {"z": 0.57, "val": 0.441, "err": 0.043},
    {"z": 0.77, "val": 0.49,  "err": 0.18},
    {"z": 1.52, "val": 0.396, "err": 0.079}
]

# ---  D(z), f(z) ---
def growth_D_full(z_array, beta, s):
    a_vals = np.linspace(1e-3, 1.0, 800)
    def dDda(Dvec, a):
        D, dD_da = Dvec
        E = np.sqrt(Omega_m * a**-3 + (1 - Omega_m))
        dlnH = -1.5 * Omega_m * a**-3 / (Omega_m * a**-3 + 1 - Omega_m)
        Ge = 1 + beta * a**(-s)
        d2D = - (3/a + dlnH) * dD_da + 1.5 * Omega_m * Ge / (a**5 * E**2) * D
        return [dD_da, d2D]
    try:
        sol = odeint(dDda, [a_vals[0], 1.0], a_vals)
        D_vals, dD_vals = sol[:,0], sol[:,1]
        if abs(D_vals[-1]) < 1e-10 or not np.isfinite(D_vals[-1]):
            return None, None
        D_vals /= D_vals[-1]
        dD_vals /= D_vals[-1]
        a_of_z = 1 / (1 + z_array)
        D_interp = interp1d(a_vals, D_vals)
        dD_interp = interp1d(a_vals, dD_vals)
        D_z = D_interp(a_of_z)
        dD_z = dD_interp(a_of_z)
        f_z = a_of_z * dD_z / D_z
        return D_z, f_z
    except:
        return None, None

# --- N(z)  ---
def sigma_M(M):
    return sigma8_0 * (M / 1e14)**-0.3

def mass_function(M, z, beta, s):
    D, _ = growth_D_full(np.array([z]), beta, s)
    if D is None or D[0] <= 0: return 0
    sigma = sigma_M(M) * D[0]
    if sigma <= 0: return 0
    delta_c, A, a, q = 1.686, 0.3222, 0.707, 0.3
    nu = delta_c / sigma
    f = A * np.sqrt(2 * a / np.pi) * nu * (1 + (nu**2 / a)**-q) * np.exp(-a * nu**2 / 2)
    return (rho_m / M) * f

def N_z(z, beta, s):
    try:
        val, _ = quad(lambda M: mass_function(M, z, beta, s), 3e14, 1e16)
        dV = 4 * np.pi * (3e3 * z)**2 * 3e3
        return val * dV
    except:
        return 0

# --- Nf8 ---
def chi2_Nz(beta, s):
    chi2 = 0
    scale = None
    for d in obs_Nz:
        N_th = N_z(d["z"], beta, s)
        if N_th <= 0: return np.inf
        if scale is None:
            scale = d["N"] / max(N_th, 1e-6)
        res = (d["N"] - scale * N_th) / d["err"]
        chi2 += res**2
    return chi2

def chi2_fs8(beta, s):
    zs = np.array([d["z"] for d in obs_fs8])
    D_z, f_z = growth_D_full(zs, beta, s)
    if D_z is None or f_z is None:
        return np.inf
    fs8_th = f_z * D_z * sigma8_0
    return sum(((d["val"] - fs8_th[i]) / d["err"])**2 for i, d in enumerate(obs_fs8))

# ---  ---
beta_vals = np.linspace(0.0, 0.15, 40)
s_vals = np.linspace(2.0, 4.0, 40)
chi2_total = np.zeros((len(s_vals), len(beta_vals)))

for i, s in enumerate(s_vals):
    for j, beta in enumerate(beta_vals):
        chi2_total[i, j] = chi2_Nz(beta, s) + chi2_fs8(beta, s)

# ---  ---
B, S = np.meshgrid(beta_vals, s_vals)
chi2_min = np.min(chi2_total)
delta_chi2 = chi2_total - chi2_min

plt.figure(figsize=(10,6))
plt.contourf(B, S, chi2_total, levels=30, cmap="inferno")
cbar = plt.colorbar()
cbar.set_label("N(z) + f(z)")

levels = [2.30, 6.17, 11.8]
CS = plt.contour(B, S, delta_chi2, levels=levels, colors='white')
plt.clabel(CS, fmt={l: f"{int(p)}%" for l,p in zip(levels, [68,95,99.7])}, fontsize=10)

i_best, j_best = np.unravel_index(np.argmin(chi2_total), chi2_total.shape)
plt.plot(beta_vals[j_best], s_vals[i_best], "c*", markersize=12, label="")

plt.xlabel("")
plt.ylabel("s")
plt.title("SPT N(z) + f(z) ")
plt.legend()
plt.tight_layout()
plt.show()
```
